```python
import numpy as np  #linear algebra
import pandas as pd   #data processing
```

```python
data =pd.read_csv("Iris.csv")
```

```python
data
```

```python
data.shape
```

Dropping the Column

```python
data=data.drop(columns=["Id"])
```

```python
data
```

```python
data.head() #Return first 5 entries
```

Transforming Categorical Data into Numeric Data

```python
data["Species"].replace({"Iris-setosa":1,"Iris-versicolor":2,"Iris-virginica":3},inplace=True)
```

```python
data
```

Creating Arrays

```python
x=pd.DataFrame(data,columns=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]).values
```

```python
x
```

```python
y=data.Species.values.reshape(-1,1)
```

```python
y
```

Importing necessary Libraries

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```


Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
k=6
knclr=KNeighborsClassifier(k)
```

Training the model using KNN (K Nearest Neighbor)

```python
knclr
```

```python
knclr.fit(x_train,y_train)
```

```python
y_pred=knclr.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_pred)*100
```
